```python
from ib_insync import *
util.startLoop()

# import nest_asyncio
# nest_asyncio.apply()


ib = IB()
# ib.disconnect()
ib.connect('127.0.0.1', 7497, clientId=3)


contract = Crypto('BTC', exchange='PAXOS', currency='USD')
bars = ib.reqHistoricalData(
    contract, endDateTime='', durationStr='1 D',
    barSizeSetting='1 min', whatToShow='MIDPOINT', useRTH=True)

# convert to pandas dataframe:
df = util.df(bars)
print(df)
```

```python
def CheckBullish(bar):
    return (bar.close > bar.open)
        
    
```

```python
# df.loc[0].close > df.loc[0].open
for i in df.iterrows():
    display(CheckBullish(i[1]))
    
```

```python
df['change']= df['close'] - df['open']
```

```python
df
```

```python
df['ten_average'] = df.close.rolling(10).mean()
df['thirty_average'] = df.close.rolling(30).mean()
df2 = df.dropna()
df2['MA_diff'] = df['thirty_average'] > df['ten_average']
df2
```

```python
def PlaceOrder(order_type):
    print(order_type)
    
def GetSignals(df):
    for count, row in enumerate(df.iterrows()):
        if count==0:
            sentiment = row[1]['MA_diff']
        else:
            if row[1]['MA_diff'] == sentiment:
                PlaceOrder('no change')
            else:
                if row[1]['MA_diff']:
                    PlaceOrder('sell')
                else:
                    PlaceOrder('buy')
                sentiment = row[1]['MA_diff']
    
```
